```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x = np.linspace(0,10,100)
x = np.linspace(0,10,4)
y = x*x

plt.figure()
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,3,10)
y1 = x
y2 = x**0.5

plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlim(0.5,3.0)
plt.ylim(0.5,3.0)
plt.title("My glorious plot")
plt.show()
```

```python
x =  np.linspace(0,1.6,30)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure()
plt.plot(y1,y2)
plt.axis('equal')
plt.show()
```

```python
x =  np.linspace(0,1.6,30)
y1 = np.sin(x)
y2 = np.cos(x)
y= np.stack([y1,y2])

plt.figure()
plt.scatter(y[0],y[1])
plt.axis('equal')
plt.show()
```

```python
x =  np.linspace(0,1.6,30)
y1 = np.sin(x)
y2 = np.cos(x)
y= np.stack([y1,y2])

y_small = y*0.3
y_shifted = y+1
y_shifted_up = y+np.array([0.0,1.0])[:,None]

plt.figure()
plt.scatter(y[0],y[1])
plt.scatter(y_small[0],y_small[1])
plt.scatter(y_shifted[0],y_shifted[1])
plt.scatter(y_shifted_up[0],y_shifted_up[1])
plt.axis('equal')
plt.show()
```
